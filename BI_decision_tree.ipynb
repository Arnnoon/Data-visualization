```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import graphviz 
from sklearn import tree
from sklearn.model_selection import train_test_split

# Reading the dataset into a pandas dataframe
df = pd.read_csv('sss.csv')

# Preprocessing the data
X = df.drop(['Timestamp', 'Frequently to visit Starbucks'], axis=1) # Dropping irrelevant columns
y = df['Frequently to visit Starbucks'] # Setting target variable

# Converting categorical data to numerical data
X = pd.get_dummies(X)

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Building and training the decision tree model
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

# Evaluating the performance of the model on the testing set
accuracy = clf.score(X_test, y_test)
print('Accuracy:', accuracy)
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=X.columns,  
                     class_names=y.unique(),  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)
graph.format = 'png'
graph.render('decision_tree')
graph
```
